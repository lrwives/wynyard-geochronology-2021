```python
# Import Packages
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xlrd

wd = os.getcwd()


# DzMix Step Relative contributions
dz_step_path = os.path.join(wd, "dz_step_realtive_contributions.csv")
dz_step = pd.read_csv(dz_step_path)
name_list = list(dz_step['sample_name'])
dz_step = dz_step.set_index('sample_name')

# Plot DzMix Step Relative contributions
fig, ax = plt.subplots()
ax.set_title('Relative Contribution of WTT Sources\nDZMix Step-Wise Model')
plt.xticks(dz_step['count'], name_list, rotation=90)

xcorr = plt.errorbar(dz_step['count'],
                     dz_step['xcorr_norm'],
                     yerr=dz_step['xcorr_std'],
                     fmt='*',
                     markersize='10',
                     color='#F0E442',
                     markeredgecolor='black',
                     ecolor='#F0E442',
                     elinewidth=1.5,
                     capsize=5,
                     label='Cross Correlation KDE')

plt.errorbar(dz_step['count'],
             dz_step['kuiper_norm'],
             yerr=dz_step['kuiper_std'],
             fmt='^',
             markersize='8',
             color='#44AA99',
             markeredgecolor='black',
             ecolor='#44AA99',
             elinewidth=1.5,
             capsize=5,
             label='Kuiper V-test')

plt.errorbar(dz_step['count'],
             dz_step['ks_norm'],
             yerr=dz_step['ks-std'],
             fmt='.',
             markersize='10',
             color='#AA4499',
             markeredgecolor='black',
             ecolor='#AA4499',
             elinewidth=1.5,
             capsize=5,
             label='K-S D-test')

# Format axes
ax.set_ylabel('Relative Contribution (%)')
ax.set_ylim(-0.05, .7)
plt.xticks(dz_step['count'], name_list, rotation=90)
plt.grid(axis='y')

# Legend
leg = ax.legend()

plt.show()
```

```python
#DzMix Non Step Relative contributions
dz_step_path = os.path.join(wd,"dz_NonStep_realtive_contributions.csv")
dz_step = pd.read_csv(dz_step_path)
name_list = list(dz_step['sample_name'])
dz_step = dz_step.set_index('sample_name')

#Plot DzMix Non Step Relative contributions
fig,ax = plt.subplots()
ax.set_title('Relative Contribution of WTT Sources\nDZMix Non Step-Wise Model')
plt.xticks(dz_step['count'],name_list, rotation=90)

xcorr = plt.errorbar(dz_step['count'],
            dz_step['xcorr_norm'],
            yerr=dz_step['xcorr_std'],
            fmt='*',
             markersize = '10',
            color='#F0E442',
             markeredgecolor='black',
            ecolor = '#F0E442',
            elinewidth = 1.5,
            capsize = 5,
            label = 'Cross Correlation KDE')

kuiper = plt.errorbar(dz_step['count'],
            dz_step['kuiper_norm'],
            yerr=dz_step['kuiper_std'],
            fmt='^',
             markersize = '8',
            color='#44AA99',
             markeredgecolor='black',
            ecolor = '#44AA99',
            elinewidth = 1.5,
            capsize = 5,
            label = 'Kuiper V-test')

ks = plt.errorbar(dz_step['count'],
            dz_step['ks_norm'],
            yerr=dz_step['ks-std'],
            fmt='.',
             markersize = '10',
            color='#AA4499',
             markeredgecolor='black',
            ecolor = '#AA4499',
            elinewidth = 1.5,
            capsize = 5,
            label = 'K-S D-test')

# Format axes
ax.set_ylabel('Relative Contribution (%)')
ax.set_ylim(-0.05,.7)
plt.xticks(dz_step['count'],name_list, rotation=90)
plt.grid(axis='y')

# Legend
leg = ax.legend()

plt.show()
```

```python
#DzMix Non Step Relative contributions
nls_path = os.path.join(wd,"nls_relative_contributions_final.csv")
nls_con = pd.read_csv(nls_path)
name_list = list(nls_con['sample_name'])
#dz_step = dz_step.set_index('sample_name')

#Plot NLS with unknowns relative contributions
fig,ax = plt.subplots()
ax.set_title('Relative Contribution of WTT Sources\nAverage of 100 Best NLS Model Fits')
plt.xticks(nls_con['count'],name_list, rotation=90)
plt.errorbar(nls_con['count'],
            nls_con['sample_norm_value'],
            yerr=nls_con['norm_std_err'],
            fmt='.',
             markersize = '10',
            color='#0072B2',
             markeredgecolor='black',
            ecolor = '#E69F00',
            elinewidth = 1.5,
            capsize = 5,
            label = 'NLS Standard Error')

# Plot NLS w/o unknowns relative contributions
plt.errorbar(nls_con['count'],
            nls_con['sample_norm_value'],
            yerr=nls_con['model_error'],
            fmt='.',
             markersize = '10',
            color='#0072B2',
             markeredgecolor='black',
            ecolor = '#0072B2',
            elinewidth = 1.5,
            capsize = 5,
            label = 'Model Uncertainty')

# Format axes
ax.set_ylabel('Relative Contribution (%)')
ax.set_ylim(-0.05,0.7)
plt.xticks(nls_con['count'],name_list, rotation=90)
plt.grid(axis='y')

# Legend
leg = ax.legend()

plt.show()
```
